```python
# Step 2: Determine the number of factors
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Standardize the features (important for PCA)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.iloc[:, 4:44])

# Create and fit the PCA model
pca = PCA()
pca.fit(df_scaled)

# Get the explained variance ratios
explained_variance_ratios = pca.explained_variance_ratio_

# Cumulative explained variance
cumulative_explained_variance = explained_variance_ratios.cumsum()

# Plot the explained variance ratios, want the factors to explain 
# very roughly 70% or more of the variance
plt.plot(range(1, len(explained_variance_ratios) + 1), \
         cumulative_explained_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()

pca.explained_variance_ # eigenvalues; want values > 1
```

# Cluster Analysis

```python

# Step 1: determine the number of clusters
# Option A: dendogram
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
dissimilarity_matrix = squareform(pdist(df.iloc[:, 52:60], metric='euclidean'))

# perform hierarchical clustering using Ward's method on this matrix
hc = linkage(dissimilarity_matrix, method="ward")

# plot the resulting dendrogram
dendrogram(hc, leaf_font_size=6)
plt.show()
```

```python
# Option B: The Elbow
from sklearn.cluster import KMeans

# Calculate the sum of squared distances for different values of k
sum_of_squared_distances = []
K_range = range(1, 11)  # You can adjust this range based on your problem
for k in K_range:
    kmeans = KMeans(n_clusters=k, n_init='auto')
    kmeans.fit(df_scaled)
    sum_of_squared_distances.append(kmeans.inertia_)

# Plot the Elbow curve
plt.plot(K_range, sum_of_squared_distances, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distances')
plt.show()
```

```python
# Step 2: Calculate the final cluster solution
kmeans = KMeans(n_clusters=5, random_state=11, n_init='auto')

# Add a new column with cluster assignment for each person
df['Cluster'] = kmeans.fit_predict(df.iloc[:, 4:44])
df[:20]
```

# Pricing Analysis

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to perform Gabor Granger analysis for all games in the dataset
def gabor_granger_analysis(df):
    games = df["gg.game.presented"].unique()  # Get all unique games

    for game in games:
        game_data = df[df["gg.game.presented"] == game]  # Filter data for the game
        
        # Get unique price points in ascending order
        price_points = sorted(game_data["gg.maxprice"].dropna().unique())

        # Initialize lists for storing results
        willingness_to_pay = []
        revenue_predictions = []

        for price in price_points:
            willing = (game_data["gg.maxprice"] >= price).sum() / len(game_data)  # % Willing to Pay
            revenue = willing * price  # Estimated Revenue

            willingness_to_pay.append(willing * 100)  # Convert to percentage
            revenue_predictions.append(revenue)

        # Convert lists to arrays
        price_points = np.array(price_points)
        willingness_to_pay = np.array(willingness_to_pay)
        revenue_predictions = np.array(revenue_predictions)

        # Plot 1: Willingness to Pay Curve
        plt.figure(figsize=(10, 5))
        plt.plot(price_points, willingness_to_pay, marker='o', linestyle='-')
        plt.xlabel("Price ($)")
        plt.ylabel("Percentage Willing to Pay (%)")
        plt.title(f"Willingness to Pay for {game}")
        plt.grid()
        plt.show()

        # Plot 2: Predicted Revenue Curve
        plt.figure(figsize=(10, 5))
        plt.plot(price_points, revenue_predictions, marker='o', linestyle='-', color='red')
        plt.xlabel("Price ($)")
        plt.ylabel("Predicted Revenue")
        plt.title(f"Predicted Revenue for {game}")
        plt.grid()
        plt.show()

        # Find the optimal price (max revenue)
        optimal_price = price_points[np.argmax(revenue_predictions)]
        print(f"Ideal price point for {game}: ${optimal_price:.2f}")

# Run the analysis
gabor_granger_analysis(df)
```

# Targeting Strategy

```python
import statsmodels.api as sm
import pandas as pd

# Ensure segment labels are categorical before encoding
df["Segment_Label"] = df["Segment_Label"].astype("category")

# One-hot encode segment labels (drop_first=True prevents multicollinearity)
df_encoded = pd.get_dummies(df, columns=["Segment_Label"], drop_first=True)

# Function to run regression for each game
def run_segment_regression(df):
    games = df["gg.game.presented"].unique()  # Get unique games

    for game in games:
        print(f"\n### Linear Regression for {game} ###\n")

        # Subset data for the game
        game_data = df[df["gg.game.presented"] == game].copy()  # Copy to avoid warnings

        # Convert gg.maxprice to numeric (if needed)
        game_data["gg.maxprice"] = pd.to_numeric(game_data["gg.maxprice"], errors="coerce")

        # Define independent variables (all segment dummies)
        X = game_data.filter(like="Segment_Label_").astype(float)  # Ensure numeric

        # Check if segment dummies exist
        if X.shape[1] == 0:
            print(f"⚠️ No segment variables found for {game}. Check encoding.")
            continue

        # Dependent variable: Max price respondents are willing to pay
        y = game_data["gg.maxprice"]

        # Ensure y is numeric
        if y.dtype == "object":
            y = pd.to_numeric(y, errors="coerce")

        # Drop NaN values
        X = X.dropna()
        y = y.loc[X.index]

        # Add constant term
        X = sm.add_constant(X)

        # Run regression
        model = sm.OLS(y, X).fit()

        # Display results
        print(model.summary())

# Run regression for each game
run_segment_regression(df_encoded)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assume df contains gg.maxprice and gg.game.presented from previous analysis
steam_users = 10_000_000  # Total potential Steam buyers
purchase_prob = 0.30  # 30% will actually buy

# Aggregate willing-to-pay data
revenue_data = {}

for game in df["gg.game.presented"].unique():
    game_data = df[df["gg.game.presented"] == game]
    
    # Group by price, count how many respondents are willing to pay
    price_wtp = game_data.groupby("gg.maxprice").size() / len(game_data)  # Willingness % per price
    
    # Estimate real buyers in Steam market
    expected_buyers = price_wtp * steam_users * purchase_prob
    
    # Calculate gross revenue
    revenue = expected_buyers * price_wtp.index  # Revenue = buyers × price
    
    # Store results
    revenue_data[game] = pd.DataFrame({
        "Price": price_wtp.index,
        "Expected Buyers": expected_buyers,
        "Gross Revenue": revenue
    })

    # Find the price that maximizes revenue
    optimal_price = revenue.idxmax()
    optimal_revenue = revenue.max()

    print(f"\n### {game} Revenue Analysis ###")
    print(f"Optimal Price: ${optimal_price:.2f}")
    print(f"Expected Gross Revenue: ${optimal_revenue:,.0f}")

    # Plot revenue curve
    plt.plot(revenue.index, revenue, marker="o", label=game)

# Show revenue plot
plt.xlabel("Price ($)")
plt.ylabel("Revenue ($)")
plt.title("Predicted Revenue for Each Game")
plt.legend()
plt.grid()
plt.show()
```

## Estimate Market Share

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of Athena's games and competitors
athena_games = ["rank.WarriorGuild", "rank.SeraphGuardians", "rank.Evercrest"]
competitor_games = ["rank.DevilsGate", "rank.Marksman", "rank.QuestoftheTitan"]
all_games = athena_games + competitor_games

# Count how many times each game was ranked 1st (most preferred)
market_share = (df[all_games] == 1).sum() / len(df) * 100  # Convert to percentage

# Display market share
print("Estimated Market Share (%) for Each Game:")
print(market_share)

# Visualizing market share
plt.figure(figsize=(8, 5))
market_share.sort_values().plot(kind="barh", color=["blue"]*3 + ["gray"]*3)
plt.xlabel("Market Share (%)")
plt.ylabel("Game")
plt.title("Estimated Market Share for Athena’s Games vs Competitors")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

```python
import pandas as pd

# Define ranking probabilities
rank_probabilities = {1: 0.8, 2: 0.6, 3: 0.4, 4: 0.2, 5: 0.1, 6: 0.05}

# Define game price mapping
game_prices = {
    "rank.WarriorGuild": 35.00,
    "rank.SeraphGuardians": 32.00,
    "rank.Evercrest": 34.00,
    "rank.DevilsGate": 14.99,
    "rank.Marksman": 9.99,
    "rank.QuestoftheTitan": 19.90
}

# Define game columns
all_games = list(game_prices.keys())

# Apply ranking probabilities to the dataset
for game in all_games:
    df[game] = df[game].map(rank_probabilities).fillna(0)  # Replace ranks with probabilities

# Calculate expected revenue per game using price
expected_revenue = {game: (df[game] * game_prices[game]).sum() for game in all_games}

# Convert to DataFrame for visualization
market_share_df = pd.DataFrame(expected_revenue.items(), columns=["Game", "Expected Revenue"])

# Normalize to calculate market share percentage
market_share_df["Market Share (%)"] = (market_share_df["Expected Revenue"] / market_share_df["Expected Revenue"].sum()) * 100

# Display the updated market share
print(market_share_df)
```

```python
# Sort market share dataframe by Market Share in descending order
market_share_df = market_share_df.sort_values(by="Market Share (%)", ascending=False)

# Plot Market Share
plt.figure(figsize=(10, 6))
plt.barh(market_share_df["Game"], market_share_df["Market Share (%)"], 
         color=['blue', 'green', 'red', 'purple', 'orange', 'cyan'])
plt.xlabel("Market Share (%)")
plt.ylabel("Game")
plt.title("Estimated Market Share Based on Rank and Price")
plt.gca().invert_yaxis()  # Invert Y-axis for better readability
plt.show()
```
